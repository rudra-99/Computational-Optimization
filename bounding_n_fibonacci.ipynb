# Fibonacci Search

```python
import matplotlib.pyplot as plt
# plt.style.use('dark_background')
```

```python
def func(x):
    return (x**2+(54/x))
```

```python
def fibo(n):
    fseries=[0,1]
    i=2
    while i<=n+1:
        fseries.append(fseries[i-1]+fseries[i-2])
        i+=1
    return fseries[n+1]
```

```python
def Fibonacci_search(a,b,n):
    low=a
    upper=b
    n_iter=n
    k=2
    L=b-a
    arr=[]
    val=[]
    while k<=n:
        Lk=(fibo(n-k+1)/fibo(n+1))*L
        x1=a+Lk
        x2=b-Lk
        arr.append([a,b])
        if func(x1)>func(x2):
            a=x1
            
        elif func(x1)<func(x2):
            b=x2
          
        elif func(x1)==func(x2):
            a,b=x1,x2
        val.append(min(func(x1),func(x2)))
        k+=1
    return [a,b],arr,val
```

```python

accuracy=[]
for i in range(3,100):
    [a,b]=Fibonacci_search(0,5,i)[0]
    accuracy.append((b-a)/(2*i))
i=range(3,100)
plt.figure(figsize=[12,10])
plt.plot(i,accuracy)
plt.xlabel('number of iterations')
plt.ylabel('accuracy')   
```

# Bounding phase

```python
def Bounding_phase(guess,increment):
    x=[guess]
    x.append(x[0]+(2**(0)*increment))
    i=2
    val=[func(x[-1])]
    while True:
        x.append(x[i-1]+(2**(i-1)*increment))
        if func(x[i])<func(x[i-1]):
            val.append(func(x[i]))
            i+=1
            
        else:
            break

    return [x[i-2],x[i]],val
```

```python
Bounding_phase(0.6,0.2)
```

```python
_,val=Bounding_phase(0.6,0.2)
plt.figure(figsize=(8,10))
plt.plot(range(len(val)),val)
```

```python

accuracy2=[]
for i in range(2,100):
    a,_=Bounding_phase(0.6,i*0.01)
    accuracy2.append((a[1]-a[0])/(2*i))
i=range(2,100)
plt.figure(figsize=[12,10])
plt.plot(i,accuracy2)
plt.xlabel('size of step')
plt.ylabel('accuracy')   
```
